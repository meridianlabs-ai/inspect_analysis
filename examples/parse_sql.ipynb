```python
from inspect_viz.data._query.parser import parse_sql

examples = [
    "SELECT name, age FROM users WHERE age > 30 ORDER BY name",
    "SELECT department, AVG(salary) as avg_salary FROM employees GROUP BY department HAVING avg_salary > 50000",
    "SELECT product_name, SUM(quantity) as total_sold FROM sales JOIN products ON sales.product_id = products.id GROUP BY product_name ORDER BY total_sold DESC LIMIT 10",
]

for i, sql in enumerate(examples):
    print(f"\nExample {i + 1}:\nSQL: {sql}")
    try:
        result = parse_sql(sql)


        print(f"Mosaic structure: {result.model_dump_json(indent=2)}")
    except Exception as e:
        print(f"Error: {e}")
```
